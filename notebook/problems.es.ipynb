## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
import random
import numpy as np
```

```python

n_simulations = 1000  # cantidad de veces que se lanzan 2 dados

# 1. simulacion de lanzar 2 dados n_simulations cantidad de veces

dice1 = np.random.randint(1, 7, n_simulations)
dice2 = np.random.randint(1, 7, n_simulations)

# 2. suma del valor total de los lanzamientos

total_scores = dice1 + dice2

# 3. definir evento de interes (suma del valor total de lanzamientos es par o superior a 7)
# numeros pares: total % 2 == 0
# superiores a 7: total > 7

event_occurrences = np.sum((total_scores % 2 == 0) | (total_scores > 7))

# 4.  estimacion de probabilidad

estimated_probability = event_occurrences / n_simulations

# 5. imprimir 
print(f"La probabilidad estimada de que la suma del valor total del lazamientos de 2 dados sea par o superior a 7: {estimated_probability:.4f}")
```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python
# 1. ball box

ball_box = {}
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif i < 30:
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

# 2. simular parametros

n_experiments = 1000
n_draws = 5

# 3. correr simulaciones

count_case1 = 0  # 3 blancas y 2 rojas
count_case2 = 0  # todas del mismo color

for _ in range(n_experiments):
    draws = [ball_box[random.randint(0, 59)] for _ in range(n_draws)]
    
    whites = draws.count("White")
    reds = draws.count("Red")
    greens = draws.count("Green")

    # (1) 3 blancas y 2 rojas
    if whites == 3 and reds == 2:
        count_case1 += 1

    # (2) todas del mismo color
    if whites == 5 or reds == 5 or greens == 5:
        count_case2 += 1

# 4 umar probabilidades

prob_case1 = count_case1 / n_experiments
prob_case2 = count_case2 / n_experiments

# 5. imprimir resultados

print(f"La probabilidad estimada de sacar 3 blancas y 2 rojas: {prob_case1:.4f}")
print(f"La probabilidad estimada de sacar todas del mismo color: {prob_case2:.4f}")
```
